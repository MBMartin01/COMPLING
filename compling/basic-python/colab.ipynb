# Running Assignment in Colab

The following is a script which lets you interact with the .py files in the assignment entirely from within colab. 

The workflow is straightforward: 

1. Mount GoogleDrive
2. Navigate to the repo (see SettingUp for how to add repo to GoogleDrive)
3. Pull any recent changes 
4. Open the py file by double-clicking it 
5. Edit file
6. Run file in colab to check result
7. Repeat 5 and 6
8. Set a commit message
9. Add files to git
10. Commit with the message
11. Push changes

The following variables should be set by you to your path (all updated in the 
first code block): 

* REPOPATH: Where you want the repo to be stored on GoogleDrive
* REPONAME: The name of the repo
* GITEMAIL: Email for github
* GITNAME: Name for github

When committing changes (Steps 8-11), change the commit message (in variable COMMITTEXT) and add the changed files. 

```python
#Should be the same as yours 
BASEPATH = "/content/drive/MyDrive/"

#Change to path to repo and repo name for you end
#For example, my repo is called Intro2PyTorch and
#is located under Projects/DavisLingTeaching/
REPOPATH = 'PATH/TO/REPO'
REPONAME = 'REPO NAME'

#Set this to your name and email for github
GITEMAIL = "YOUREMAIL@email.com"
GITNAME = "YOUR NAME"
```
